```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def step_function(x):
    return np.array(x > 0, dtype=np.int)
```

```python
x = np.arange(-5.0,5.0,0.1)
y = step_function(x)
```

```python
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.array([-1.0, 1.0, 2.0])
y = sigmoid(x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
y1 = step_function(x)
y2 = sigmoid(x)
```

```python
plt.plot(x, y2, label="sigmoid")
plt.plot(x,y1, linestyle="--", label = "step_function")
plt.title('sigmid&step_funtion')
plt.legend()
plt.show()
```
